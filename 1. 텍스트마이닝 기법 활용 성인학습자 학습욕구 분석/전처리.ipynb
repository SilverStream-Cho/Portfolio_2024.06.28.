```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 엑셀 파일 경로 설정
excel_file_path = '알라딘_베스트_국내도서_2019년.xlsx'

# 엑셀 파일을 데이터프레임으로 불러오기
df = pd.read_excel(excel_file_path)

# 선택한 열 데이터 전처리
selected_columns = ['출판사/제작사', '정가', '분야', '순위', '출간연도', '출간월']
selected_df = df[selected_columns]

# '정가' 컬럼을 숫자형으로 변환
selected_df['정가'] = pd.to_numeric(selected_df['정가'], errors='coerce')

# '정가' 컬럼의 NaN 값을 0으로 대체
selected_df['정가'].fillna(0, inplace=True)

# '출판사' 컬럼의 NaN 값을 'Unknown'으로 대체
selected_df['출판사/제작사'].fillna('Unknown', inplace=True)

# TF-IDF 벡터화 (출판사 열에 대해서만 진행)
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(selected_df['출판사/제작사'].astype(str))

# 표준화 (정가 컬럼에 대해서만 진행)
scaler = StandardScaler()
selected_df[['정가']] = scaler.fit_transform(selected_df[['정가']])

# k-means 클러스터링 수행
k = 4
kmeans = KMeans(n_clusters=k, random_state=42)
selected_df['클러스터'] = kmeans.fit_predict(tfidf_matrix)

# 속성별 히스토그램 그리기
attributes = ['정가', '분야', '순위', '출간연도', '출간월']

# 클러스터별 색깔 리스트 정의
cluster_colors = ['blue', 'orange', 'green', 'red']

for attribute in attributes:
    plt.figure(figsize=(10, 6))
    for cluster in range(k):
        cluster_data = selected_df[selected_df['클러스터'] == cluster]
        sns.histplot(data=cluster_data, x=attribute, color=cluster_colors[cluster], label=f'Cluster {cluster}', kde=True)
    plt.title(f'{attribute} 히스토그램 with Clusters')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram
import matplotlib.pyplot as plt

# 엑셀 파일 경로 설정
excel_file_path = '알라딘_베스트_국내도서_2019년.xlsx'

# 엑셀 파일을 데이터프레임으로 불러오기
df = pd.read_excel(excel_file_path)

# 선택한 열 데이터 전처리
selected_columns = ['출판사', '정가', '분야', '순위', '출간연도', '출간월']
selected_df = df[selected_columns]

# '정가' 컬럼을 숫자형으로 변환
selected_df['정가'] = pd.to_numeric(selected_df['정가'], errors='coerce')

# '정가' 컬럼의 NaN 값을 0으로 대체
selected_df['정가'].fillna(0, inplace=True)

# '출판사' 컬럼의 NaN 값을 'Unknown'으로 대체
selected_df['출판사'].fillna('Unknown', inplace=True)

# TF-IDF 벡터화 (출판사 열에 대해서만 진행)
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(selected_df['출판사'].astype(str))

# 계층적 클러스터링 수행
linkage_matrix = linkage(tfidf_matrix.toarray(), method='ward')  # Ward linkage 사용
max_d = 2.5  # 클러스터 간 최대 거리 설정
clusters = fcluster(linkage_matrix, max_d, criterion='distance')

# 클러스터링 결과를 데이터프레임에 추가
selected_df['클러스터'] = clusters

# 각 클러스터별 개수 출력
cluster_counts = selected_df['클러스터'].value_counts().sort_index()
for cluster, count in cluster_counts.items():
    print(f"클러스터 {cluster}: {count}개")

# 일부 샘플만 사용하여 덴드로그램 시각화
sample_indices = range(0, min(100, selected_df.shape[0]))  # 최대 100개 샘플 사용
sample_linkage_matrix = linkage_matrix[sample_indices]
plt.figure(figsize=(10, 5))
dendrogram(sample_linkage_matrix)
plt.xlabel('샘플 번호')
plt.ylabel('클러스터 거리')
plt.title('계층적 클러스터링 덴드로그램 (일부 샘플)')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram
import matplotlib.pyplot as plt

# 엑셀 파일 경로 설정
excel_file_path = '알라딘_베스트_국내도서_2019년.xlsx'

# 엑셀 파일을 데이터프레임으로 불러오기
df = pd.read_excel(excel_file_path)

# 선택한 열 데이터 전처리
selected_columns = ['정가', '분야 번호', '순위', '출간연도', '출간월']
selected_df = df[selected_columns]

# '정가' 컬럼의 NaN 값을 0으로 대체
selected_df['정가'].fillna(0, inplace=True)

# '분야 번호', '순위', '출간연도', '출간월' 컬럼의 NaN 값을 'Unknown'으로 대체
selected_df['분야 번호'].fillna('Unknown', inplace=True)
selected_df['순위'].fillna('Unknown', inplace=True)
selected_df['출간연도'].fillna('Unknown', inplace=True)
selected_df['출간월'].fillna('Unknown', inplace=True)

# 범주형 데이터를 숫자형으로 변환
category_columns = ['분야 번호', '순위', '출간연도', '출간월']
for col in category_columns:
    selected_df[col] = pd.factorize(selected_df[col])[0]

# 표준화 (정가 컬럼에 대해서만 진행)
scaler = StandardScaler()
selected_df[['정가']] = scaler.fit_transform(selected_df[['정가']])

# 계층적 클러스터링 수행
linkage_matrix = linkage(selected_df, method='ward')  # Ward linkage 사용
t = 10  # 클러스터 개수에 따라 임계값 설정
clusters = fcluster(linkage_matrix, t, criterion='maxclust')

# 클러스터링 결과를 데이터프레임에 추가
selected_df['클러스터'] = clusters

# 각 클러스터별 개수 출력
cluster_counts = selected_df['클러스터'].value_counts().sort_index()
for cluster, count in cluster_counts.items():
    print(f"클러스터 {cluster}: {count}개")

# 덴드로그램 시각화 (옵션)
plt.figure(figsize=(10, 5))
dendrogram(linkage_matrix)
plt.xlabel('샘플 번호')
plt.ylabel('클러스터 거리')
plt.title('계층적 클러스터링 덴드로그램')
plt.show()
```
