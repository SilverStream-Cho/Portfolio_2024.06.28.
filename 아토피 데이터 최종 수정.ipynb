```python
import pandas as pd

# 파일 경로 설정 (파일이 현재 작업 디렉토리에 있을 경우 파일 이름만 사용해도 됩니다)
file_path = "data/2017-2019_서울및6개광역시_아토피_지역명수정.xlsx"

# CSV 파일을 데이터프레임으로 읽기
df1 = pd.read_excel(file_path) 
df1
```

```python
df1.rename(columns={'지역명': '지역'}, inplace=True)
```
